```python
# prompt: i Want to retrieve  data from this excel.
# Here "A" contains employee id , "B" Contains d/m/y format , "C" contains employees all login time of the day. Each entry represents a single login. one employee logged in multiple times. so the ID column does not contain unique entity. I want to retrieve the data of all employees first and last login time of a day

import pandas as pd

# Assuming 'df' is already loaded as in the preceding code

# Convert column 'B' to datetime objects
df['B'] = pd.to_datetime(df['B'], dayfirst=True)

# Extract date from datetime
df['Date'] = df['B'].dt.date

# Group by employee ID and date to find first and last login times
result = df.groupby(['A', 'Date'])['B'].agg(['min', 'max'])

# Rename columns for clarity
result.columns = ['First Login', 'Last Login']

# Reset index to make 'A' and 'Date' regular columns
result = result.reset_index()

print(result)
```

```python
# prompt: print result in a text file

with open('result.txt', 'w') as f:
  f.write(result.to_string())
```
